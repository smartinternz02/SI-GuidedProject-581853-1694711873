```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/House Price India.csv')
df.head()
```

```python
#univariate analysis
summary_stats = df['Price'].describe()
summary_stats
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Price'], kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Price')
plt.show()
```

```python
#bi-variate analysis
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Price', y='number of bedrooms')
plt.xlabel('Price')
plt.ylabel('number of bedrooms')
plt.title('Scatter Plot of Price vs. number od bedrooms')
plt.show()
```

```python
#multi variate analysis
sns.pairplot(df[['Price', 'number of bedrooms', 'lot area', 'condition of the house','Built Year','lot_area_renov']])
plt.show()
```
